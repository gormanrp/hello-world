## Device Data Resource Tree

```python
dot = Digraph(comment='Resource Tree', format='png')

dot.graph_attr['rankdir'] = 'LR'
dot.node_attr['shape'] = 'box'
dot.graph_attr['labelloc'] = "top"
dot.graph_attr['label'] = "Sample Device Data Resource Tree"

dot.node("title1", "CSE", shape="underline")
dot.node("title2", "The Car is the device", shape="underline")
dot.node("title3", "Two temperature modules in the car", shape="underline")
dot.node("title4", "Two temperature sensors per module", shape="underline")
dot.node("title5", "Historical temperature readings for each sensor", shape="underline")

dot.edge("title1","title2", style="invis")
dot.edge("title2","title3", style="invis")
dot.edge("title3","title4", style="invis")
dot.edge("title4","title5", style="invis")

dot.node("cse", "Chordant-IN-CSE")
dot.node("deviceNode", "'myCarNode'")
dot.node("deviceAE", "'myCarAE'")
dot.node("objectTitle","Flex Containers\nObject ID_Object Instance #\n(non-Data resources)",shape="none")
dot.node("resourceTitle","Regular Containers (or Time Series)\nResource Instance #\n(of Data resource)",shape="none")
dot.node("dataTitle","Content Instances (or Time Series Instances)",shape="none")
dot.node("19_0")
dot.node("19_1")
dot.node("19_0_0", "0")
dot.node("19_0_1", "1")
dot.node("19_1_0", "0")
dot.node("19_1_1", "1")
dot.node("19_0_0_0", " <auto-generated-name>")
dot.node("19_0_0_1", " <auto-generated-name>")
dot.node("19_0_1_0", " <auto-generated-name>")
dot.node("19_0_1_1", " <auto-generated-name>")
dot.node("19_1_0_0", " <auto-generated-name>")
dot.node("19_1_0_1", " <auto-generated-name>")
dot.node("19_1_1_0", " <auto-generated-name>")
dot.node("19_1_1_1", " <auto-generated-name>")

dot.edge("cse","deviceAE")
dot.edge("cse","deviceNode")
dot.edges([("19_0","19_0_0"),("19_0","19_0_1")]) # Just doing edges a different way instead of one at a time.
dot.edges([("19_1","19_1_0"),("19_1","19_1_1")])
dot.edge("19_0","resourceTitle",style="invis")
dot.edge("deviceAE","objectTitle",style="invis")
dot.edge("deviceAE","19_0")
dot.edge("deviceAE","19_1")

dot.edge("19_0_0","dataTitle",style="invis")
dot.edge("19_0_0","19_0_0_0")
dot.edge("19_0_0","19_0_0_1")
dot.edge("19_0_1","19_0_1_0")
dot.edge("19_0_1","19_0_1_1")
dot.edge("19_1_0","19_1_0_0")
dot.edge("19_1_0","19_1_0_1")
dot.edge("19_1_1","19_1_1_0")
dot.edge("19_1_1","19_1_1_1")

dot.render('tree.gv', view=False) # Creates output file (as per format set above). Can't remember where I got the .gv extension from, but it's a .dot file

dot
```

## Device Data Link Attribute

```python
dot = Digraph(comment='Link', format='png')

dot.graph_attr['rankdir'] = 'LR'
dot.node_attr['shape'] = 'box'
dot.graph_attr['labelloc'] = "top"
dot.graph_attr['label'] = "Device Data Link Attribute"

dot.node("deviceNode", " <device node>")
#dot.node("title", "Pertinent Attribute", shape="underline")
dot.node("link", "hael", shape="oval")

#dot.edge("deviceNode","title",style="invis")
dot.edge("deviceNode","link")

dot.node("linkNote","Hosted AE Links: Optional: m2m:listOfM2MID\lSet to the one AE ID that is the parent of the Object 19, if any",shape="none",fontcolor="red")

dot.edge("link","linkNote",style="invis")

dot.render('link.gv', view=False) # Creates output file (as per format set above). Can't remember where I got the .gv extension from, but it's a .dot file

dot
```

## Device Data Attributes

```python
dot = Digraph(comment='Attributes', format='png')

dot.graph_attr['rankdir'] = 'LR'
dot.node_attr['shape'] = 'box'
dot.graph_attr['labelloc'] = "top"
dot.graph_attr['label'] = "Device Data Attributes"

dot.node("flex", "19_0")
dot.node("title", "Flex Container\nCustom Attributes\n(Meta Data)", shape="underline")
dot.node("priority", "1", shape="oval")
dot.node("time", "2", shape="oval")
dot.node("descr", "3", shape="oval")
dot.node("format", "4", shape="oval")
dot.node("app", "5", shape="oval")

dot.edge("flex","title",style="invis")
dot.edge("flex","priority")
dot.edge("flex","time")
dot.edge("flex","descr")
dot.edge("flex","format")
dot.edge("flex","app")

dot.node("priorityNote","Data Priority: Optional: Integer",shape="none")
dot.node("timeNote","Data Creation Time: Optional: Time\lPresence indicates Time Series",shape="none")
dot.node("descrNote","Data Description: Optional: String",shape="none")
dot.node("formatNote","Data Format: Optional: String",shape="none")
dot.node("appNote","App ID: Optional: Integer: Destination application ID",shape="none")

dot.edge("priority","priorityNote",style="invis")
dot.edge("time","timeNote",style="invis")
dot.edge("descr","descrNote",style="invis")
dot.edge("format","formatNote",style="invis")
dot.edge("app","appNote",style="invis")

dot.render('attributes.gv', view=False) # Creates output file (as per format set above). Can't remember where I got the .gv extension from, but it's a .dot file

dot
```
